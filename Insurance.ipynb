```python
#encoding the catogerical features
#sex columns

df.replace({'sex':{'male':0,"female":1}},inplace=True)

#smoker column
df.replace({'smoker':{'yes':0,'no':1}},inplace=True)

#region column
df.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
```

```python
df
```

```python
x=df.drop(columns='charges',axis=1)
y=df['charges']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
regrissior=LinearRegression()
regrissior.fit(x_train,y_train)
```

```python
#prediction on training data

training_data_prediction=regrissior.predict(x_train)
# R-squared value for training data
r2_train = metrics.r2_score(y_train, training_data_prediction)

# Prediction on testing data
testing_data_prediction = regrissior.predict(x_test)

# R-squared value for testing data
r2_test = metrics.r2_score(y_test, testing_data_prediction)

# Print R-squared values
print(f"R-squared value for training data: {r2_train}")
print(f"R-squared value for testing data: {r2_test}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_regression = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regression.fit(x_train, y_train)
rf_regression.fit(x_test, y_test)
```

```python
y_pred = rf_regression.predict(x_train)
y_pred1 = rf_regression.predict(x_test)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
svm=SVR()
svm.fit(x_train,y_train)
train=svm.predict(x_train)
test=svm.predict(x_test)
print('train accuracy',r2_score(train,y_train))
print('test accuracy',r2_score(test,y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
train=knn.predict(x_train)
test=knn.predict(x_test)
print('train accuracy',r2_score(train,y_train))
print('test accuracy',r2_score(test,y_test))
```
